##Example of more buliful bar plot

```python
# Set the context and style for better aesthetics
sns.set_context('talk')  # Makes the font larger and clearer
sns.set_style('whitegrid')  # Use a white background with grid lines

# Create a bar plot with a more vibrant palette and borders
plt.figure(figsize=(10, 8))
ax = sns.barplot(x=Top_5_cat_wrt_rating.index, y=Top_5_cat_wrt_rating.values,
                 palette='magma', edgecolor='black', linewidth=1.5)

# Add labels and title with larger fonts and styling
plt.title('Top 10 Categories by Count', fontsize=18, weight='bold', color='darkblue', pad=20)
plt.xlabel('Category', fontsize=14, weight='bold')
plt.ylabel('Count', fontsize=14, weight='bold')

# Add labels to each bar with a shadow effect for better readability
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}',
                (p.get_x() + p.get_width() / 2, p.get_height()),
                ha='center', va='center',
                xytext=(0, 10),  # Offset label to avoid overlap
                textcoords='offset points', fontsize=12, weight='bold',
                color='black', bbox=dict(facecolor='white', alpha=0.8, boxstyle='round,pad=0.5'))

# Rotate the x-axis labels for better readability and increase font size
plt.xticks(rotation=45, ha='right', fontsize=12, weight='bold')

# Remove the top and right spines for a cleaner look
sns.despine()

# Adjust the layout and add some space between elements
plt.tight_layout(pad=2)

# Show the plot
plt.show()
```

##What are the Top 5 most installed apps in Each popular Cat

```python
cdf.head()
```

```python
Sorted_cat=cdf.groupby(['Category','App'])['Installs'].sum().reset_index()
```

```python
Sorted_cat=Sorted_cat.sort_values('Installs',ascending=False)
```

```python
Sorted_cat[Sorted_cat.Category=='GAME'][:5]
```

```python
Sorted_cat[Sorted_cat.Category=='FAMILY'][:5]
```

```python
Sorted_cat[Sorted_cat.Category=='PRODUCTIVITY'][:5]
```

```python
# List of categories to filter
categories = ['GAME', 'FAMILY', 'TOOLS','PRODUCTIVITY']

# Loop over each category and filter the data
for category in categories:
    # Filter the DataFrame for the given category and select the top 5
    filtered_data = Sorted_cat[Sorted_cat.Category == category][:5]

    # Print the results (or you can store them in a list or dictionary if needed)
    print(f"Top 5 entries for category: {category}")
    print(filtered_data)
    print("\n")  # Adds space between outputs for readability
```
