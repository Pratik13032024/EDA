###Converting Date_of_Journey colum type into datetime from object and extracting the Day, Month & Year column

```python
#Coverting object format to datetime
df['Date_of_Journey']=pd.to_datetime(df['Date_of_Journey'],infer_datetime_format=True)
```

```python
df.info()
```

```python
df['Day']=df['Date_of_Journey'].dt.day
df['Month']=df['Date_of_Journey'].dt.month
df['Year']=df['Date_of_Journey'].dt.year
```

```python
df.head()
```

```python
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df.head()
```

###Dep_time time converting from object to datetime format and adding two seprate columns as dpt_hr and dpt_min

```python
df.head()
```

```python
df['Dep_Time']=pd.to_datetime(df['Dep_Time'],infer_datetime_format=True)
```

```python
df.info()
```

```python
df['Dept_Hour']=df['Dep_Time'].dt.hour
df['Dept_min']=df['Dep_Time'].dt.minute
```

```python
df.head()
```

```python
df.drop('Dep_Time',axis=1,inplace=True)
```

```python
df.head()
```

```python
df.info()
```

###Arrival_Time converting from object to datetime format and adding two seprate columns as Arrival_hour and Arrival_min

```python
df.head()
```

```python
df['Arrival_Time']=df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
df['Arrival_Time']
```

```python
df['Arrival_hour']=df['Arrival_Time'].str.split(':').str[0]
df['Arrival_min']=df['Arrival_Time'].str.split(':').str[1]
```

```python
df['Arrival_hour']=df['Arrival_hour'].astype(int)
df['Arrival_min']=df['Arrival_min'].astype(int)
```

```python
df.head()
```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df.info()
```

###Total_Stops values into int I have decidec to convert non-stop : 0 , 1 stop:1 , 2 stop:2 , 3 stop:3 , 4 stop:4 and for this we have to map dict

```python
df.head()
```

```python
df['Total_Stops'].unique()
```

```python
df['Total_Stops'].value_counts()
```

```python
df[df['Total_Stops'].isnull()]
```

```python
#Now to convert Total stop values into int I have decidec to convert non-stop : 0 , 1 stop:1 , 2 stop:2 , 3 stop:3 , 4 stop:4 and for this we have to map dict

df['Total_Stops']=df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,'nan':1})
```
