```python
# To change index name #
d3 = {'k1':[1,2,3,4,5],
      'k2':[6,7,8,9,10],
      'k3':[11,12,13,14,15],}
```

```python
df1=pd.DataFrame(d3 )
```

```python
df1
```

```python
d4 = {'k1':[1,2,3,4,5,'pratik'],
      'k2':[6,7,8,'kumar',9,10],
      'k3':[11,12,13,14,15,'dombale']}
```

```python
df2=pd.DataFrame(d4)
```

```python
df2
```

```python
## Marge operation ##
```

```python
pd.merge(df1,df2)## Only sow  common parts
```

```python
pd.merge(df1,df2,how= 'inner',on='k1')### meaning of 'on'----it will match withrespect to k1
```

```python
pd.merge(df1,df2,how= 'inner',on='k3')
```

```python
pd.merge(df1,df2,how= 'left',on='k3')#--- bring all data from left hand side i.e df1
```

```python
pd.merge(df1,df2,how= 'outer',on='k3')#----- bring every combination from left and right
```
