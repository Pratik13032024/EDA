```python
smartphone=df[df['Product Type'] == 'Smartphone']
smartphone
```

```python
product_wise_distribution=smartphone.groupby('Gender').size().reset_index(name='Gender wise distribution')
product_wise_distribution
```

```python
plt.figure(figsize=(3,3))
sns.barplot(x='Gender',y='Gender wise distribution' , data=product_wise_distribution)
plt.title('Total Count of Smartphone by Gender Wise',fontsize = 9)
plt.xlabel('Gender',fontsize=9)
plt.ylabel('Number of Smartphone',fontsize=9)
```
